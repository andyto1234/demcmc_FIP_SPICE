```python
import xarray as xr
from astropy import units as u


# Retrieve dimensions of the solar map
y_dim = sunpy.map.Map(files[0])[0].dimensions[1].value
x_dim = sunpy.map.Map(files[0])[0].dimensions[0].value
num_files = len(files)

wcs_original = sunpy.map.Map(files[0])[0].wcs

# Create a blank cube to store solar map data
blank = np.zeros((int(y_dim), int(x_dim), num_files,))

# Populate the blank cube with data from each file
for num, file in enumerate(files):
    blank[:, :, num] = sunpy.map.Map(file)[0].data

# Extract wavelengths from the files
wavelengths = [sunpy.map.Map(file)[0].meta['wavelength'] for file in files]
ions = [sunpy.map.Map(file)[0].meta['ion'] for file in files]
combined_list = []

for ion, wavelength in zip(ions, wavelengths):
    combined_list.append(f"{ion}_{wavelength:.2f}")

print(combined_list)
```

```python
# Create an xarray Dataset
dataset = xr.Dataset(
    data_vars=dict(
        data=(["y", "x", "wavelength"], blank)
    ),
    coords=dict(
        y=np.arange(y_dim),
        x=np.arange(x_dim),
        wavelength=wavelengths,
        linenames=combined_list,
    ),
    attrs=dict(
        description="SPICE map data"
    )
)

# Convert WCS object to a dictionary and add it as an attribute
dataset.attrs["wcs"] = wcs_original.to_header_string()

# Save the dataset to a file (optional)
dataset.to_netcdf("solar_data.nc")
```

```python
def get_density(obs_ratio, dens_sav):
    from scipy.io import readsav
    import numpy as np
    from astropy.visualization import ImageNormalize
    import astropy.units as u

    density_ratios = readsav(dens_sav)['smooth_rat']
    density_values = readsav(dens_sav)['smooth_den']

    # Reshape obs_ratio to a 1D array
    obs_ratio_1d = obs_ratio.ravel()

    # Compute the absolute difference between density_ratios and each value in obs_ratio_1d
    diff = np.abs(density_ratios - obs_ratio_1d[:, np.newaxis])

    # Find the indices of the minimum values along the second axis
    closest_indices = np.argmin(diff, axis=1)

    # Get the corresponding density values using the closest indices
    obs_ratio_1d = density_values[closest_indices]

    # Reshape the result back to the original shape of obs_ratio
    obs_ratio = obs_ratio_1d.reshape(obs_ratio.shape)

    return obs_ratio
```

```python
from astropy.modeling import models
from gwcs import coordinate_frames as cf
from gwcs import wcs as gwcs
from astropy import units as u

# Retrieve dimensions of the solar map
y_dim = sunpy.map.Map(files[0])[0].dimensions[1].value
x_dim = sunpy.map.Map(files[0])[0].dimensions[0].value
num_files = len(files)
wcs_original = sunpy.map.Map(files[0])[0].wcs

# Create a blank cube to store solar map data
blank = np.zeros((num_files, int(y_dim), int(x_dim), ))
# Populate the blank cube with data from each file
for num, file in enumerate(files):
    blank[num,:, :] = sunpy.map.Map(file)[0].data
    
wavelengths = [sunpy.map.Map(file)[0].meta['wavelength'] for file in files]

# Create a custom transformation for wavelengths
wavelength_transform = models.Tabular1D(points=np.arange(len(wavelengths)), lookup_table=wavelengths, bounds_error=False, fill_value=wavelengths[-1])

# Create coordinate frames
sky_frame = cf.CelestialFrame(axes_order=(0, 1), name='helioprojective', reference_frame=wcs_original.celestial.frame)
spectral_frame = cf.SpectralFrame(axes_order=(2,), unit=u.Angstrom, axes_names=('wavelength',))

# Create the WCS object
wcs = gwcs.WCS(forward_transform=wavelength_transform, input_frame=cf.CoordinateFrame(name='pixel', naxes=3),
               output_frame=[sky_frame, spectral_frame])

# Create the ndcube
cube = ndcube.NDCube(data=blank, wcs=wcs)
```

```python
wcs = astropy.wcs.WCS(naxis=3)
wcs.wcs.ctype = 'HPLT-TAN', 'HPLN-TAN', "WAVE"
wcs.wcs.cunit = wcs_original.wcs.cunit[0], wcs_original.wcs.cunit[1], 'Angstrom'
wcs.wcs.cdelt = wcs_original.wcs.cdelt[0], wcs_original.wcs.cdelt[1], 1
wcs.wcs.crpix = wcs_original.wcs.crpix[0], wcs_original.wcs.crpix[1], 1
wcs.wcs.crval = wcs_original.wcs.crval[0],wcs_original.wcs.crval[1], wavelengths[0]
wcs.wcs.cname = '', '', 'wavelength'

# Set the wavelength values for each slice
wcs.wcs.crval[2] = wavelengths[0]
wcs.wcs.pc = np.eye(3)
wcs.wcs.pc[:2, :2] = wcs_original.wcs.pc
wcs.wcs.cdelt[2] = np.diff(wavelengths).mean()

# Create an NDCube from the data and its WCS
example_cube = NDCube(blank, wcs=wcs)
```
